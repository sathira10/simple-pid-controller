# Simulation of PID controller

```python
import control

from controller import SimplePIDController
from simulation import simulate_controller
```

```python
# define transfer function
G = control.TransferFunction(1.44, [1.24e-5, 0.000746, 0.00327]) 
dt = 0.001 # sample time
ref = 50 # reference speed
```

### P only control

```python
# PID gains
kp = 0.05
ki = 0
kd = 0

# create an instance of the PID controller
pid = SimplePIDController(kp=kp, ki=ki, kd=kd, lim_min=0, lim_max=0.95, T=dt)

# simulation
y, pwm, t = simulate_controller(G, dt, pid, ref)
```

### PI control
P only control results in a steady state error. I term is introduced to remedy this.

```python
# PID gains
kp = 0.05
ki = 0.1
kd = 0

# create an instance of the PID controller
pid = SimplePIDController(kp=kp, ki=ki, kd=kd, lim_min=0, lim_max=0.95, T=dt)

# simulation
y, pwm, t = simulate_controller(G, dt, pid, ref)
```

### PID control
However the transient response is not satisfactory. Therefore, we introduce a D term. Parameters were tweaked until a satisfactory response was obtained.

```python
# PID gains
kp = 0.1
ki = 0.5
kd = 0.004

# create an instance of the PID controller
pid = SimplePIDController(kp=kp, ki=ki, kd=kd, lim_min=0, lim_max=0.95, T=dt)

# simulation
y, pwm, t = simulate_controller(G, dt, pid, ref)
```
